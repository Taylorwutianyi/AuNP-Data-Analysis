```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('npsize.txt', skiprows=6, sep=' ', index_col=0)
df.head()
```

```python
df.drop(columns=['Unnamed: 3'], inplace=True)
df
```

```python
g = sns.JointGrid(x="Length", y="Width", data=df)

ax = sns.kdeplot(data=df, x='Length', y='Width', cmap='Blues', ax=g.ax_joint)
sns.kdeplot(x=df['Length'], ax=g.ax_marg_x, linewidth=3)
sns.kdeplot(y=df['Width'], ax=g.ax_marg_y, linewidth=3)

ax.set_xlabel('Length (nm)', fontdict={'fontsize': 20}, c='k')
ax.set_ylabel('Width (nm)', fontdict={'fontsize': 20}, c='k')
ax.tick_params(axis='both', labelsize=15, colors='k')
g.savefig("1.png", bbox_inches='tight')
```
